```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import wget
```

```python
def check_generator(generator):
  """
  Фунцкия проверки работы генератора. Отрисовывает 20
  сгенерированных изображений
  """
  random_latent_vectors = tf.random.normal(shape=(20, 128))
  generated_images = generator(random_latent_vectors)
  generated_images *= 255
  generated_images.numpy()
  plt.figure(figsize=(12, 8))
  for i in range(20):
    img = keras.utils.array_to_img(generated_images[i])
    plt.subplot(4, 5, i+1)
    plt.imshow(img)
    plt.axis('off')
  plt.show()
```

```python
# считываем модель
generator = load_model('/content/best_gen')
```

```python
# проверяем работу, можно перезапустить для разных результатов
check_generator(generator)
```
